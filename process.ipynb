Importing pandas

```python
import pandas as pd
```

Import data set

```python
df = pd.read_csv('dataset.csv')
print(df)
```

```python
print(df.head())
```

```python
print(df.columns)
```

```python
df_encoded = df.copy()
```

```python
df.info()
```

```python
print(df['GENDER'])
```

```python
df_encoded['GENDER'] = df_encoded['GENDER'].map({'M':0,'F':1})
```

```python
df_encoded.head()
```

```python
df_encoded['LUNG_CANCER']=df_encoded['LUNG_CANCER'].map({'YES':1, 'NO':0})
```

```python
df_encoded.head()
```

Split data set into features(X) and target(Y), in our case, LUNG_CANCER column is the target and others are features

```python
X = df_encoded.drop('LUNG_CANCER', axis=1)
```

```python
y = df_encoded['LUNG_CANCER']
```

Import Library

```python
from sklearn.model_selection import train_test_split
```

split train test

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

Training the model

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

Prediction

```python
y_pred = model.predict(X_test)
```

```python
Input_data = [
  0,  
  55, 
  1,  
  2,  
  2,  
  1,  
  1,  
  2,  
  1,  
  2, 
  2,  
  2, 
  2,  
  2,  
  2 
]
prediction = model.predict([Input_data])

if prediction ==1:
  print("Prediction is: LUNG CANCER (YES)")
else:
  print("Prediction is: LUNG CANCER (NO)")
```

Evaluating the model

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy:{accuracy:.4f}")

print("classification_report: ")
print(classification_report(y_test, y_pred))

conf_matrix = confusion_matrix(y_test, y_pred)
print("confusion_matrix: ")
print(conf_matrix)

sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Cancer', 'Cancer'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel("Actual")
plt.show()
```
